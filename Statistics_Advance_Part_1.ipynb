#  Statistics Advance Part 1

---

## 1. What is a random variable in probability theory?

A **random variable** is a function that assigns a **numerical value** to each outcome in the sample space of a random experiment.

**Example**: Tossing a coin  
Let \( X \) be the random variable:  
- \( X = 1 \) if heads  
- \( X = 0 \) if tails  

---

## 2. What are the types of random variables?

1. **Discrete Random Variable**: Takes countable values (like 0, 1, 2...).  
2. **Continuous Random Variable**: Takes values in a continuous interval.

**Examples**:  
- Discrete: Number of students in a class  
- Continuous: Weight of a person

---

## 3. Difference between discrete and continuous distributions

| Feature                   | Discrete Distribution       | Continuous Distribution       |
|---------------------------|-----------------------------|-------------------------------|
| Values Taken              | Countable                   | Uncountable                   |
| Probability of Exact Value| Non-zero                    | Always zero                   |
| Examples                  | Binomial, Poisson           | Normal, Exponential           |
| Function Used             | PMF                         | PDF                           |

---

## 4. What are probability distribution functions (PDF)?

A **PDF** describes the **likelihood** for a continuous random variable to take a specific value.  
- For continuous variables only  
- Area under curve = 1  
- \( f(x) \geq 0 \)

---

## 5. How do CDFs differ from PDFs?

| Feature      | PDF                               | CDF                               |
|--------------|------------------------------------|------------------------------------|
| Full Form    | Probability Density Function       | Cumulative Distribution Function   |
| Purpose      | Likelihood at a point              | Probability up to a point          |
| Formula      | \( f(x) = \frac{d}{dx}F(x) \)      | \( F(x) = P(X \le x) \)            |
| Application  | Continuous variables               | All types of variables             |

---

## 6. What is a discrete uniform distribution?

- Equal probabilities to all outcomes in a finite set  
**Example**: Fair dice  
\[
P(X = x) = \frac{1}{n}, \quad x \in \{1, 2, ..., n\}
\]

---

## 7. Key properties of a Bernoulli distribution

- Two outcomes: 1 (success), 0 (failure)  
- PMF:  
\[
P(X = x) = p^x (1 - p)^{1 - x}
\]  
- Mean: \( p \)  
- Variance: \( p(1 - p) \)

---

## 8. What is the binomial distribution?

- Represents number of successes in \( n \) trials  
- PMF:  
\[
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
\]  
- Use: Tossing coins, quality control

---

## 9. What is the Poisson distribution?

- Models rare events over time/space  
- PMF:  
\[
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
\]  
- Example: Calls per hour in a call center

---

## 10. What is a continuous uniform distribution?

- Equal likelihood over interval \([a, b]\)  
- PDF:  
\[
f(x) = \frac{1}{b - a}, \quad a \le x \le b
\]  
- Example: Random number between 0 and 1

---

## 11. Characteristics of a normal distribution

- Bell-shaped, symmetric  
- Defined by:  
  - Mean \( \mu \)  
  - Standard deviation \( \sigma \)  
- Mean = Median = Mode  
- Applications: Heights, weights, IQ scores

---

## 12. What is the standard normal distribution?

- A normal distribution with  
  - Mean \( \mu = 0 \)  
  - Std dev \( \sigma = 1 \)  
- Used to calculate **Z-scores**  
- Basis for hypothesis testing and confidence intervals

---

## 13. What is the Central Limit Theorem (CLT)?

> As sample size increases, the distribution of sample means approaches normality regardless of the original population distribution.

- Enables use of normal distribution for inference  
- \(\bar{X} \sim N(\mu, \sigma^2/n)\) as \(n \to \infty\)

---

## 14. CLT and its relation to normal distribution

- Even if the population isn't normal, the sample mean will be  
- Larger sample size = better approximation

---

## 15. Application of Z statistics in hypothesis testing

- Used when population standard deviation is known  
- Formula:  
\[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
\]  
- Compare Z to standard normal critical values

---

## 16. How to calculate a Z-score?

\[
Z = \frac{X - \mu}{\sigma}
\]  
- Represents how many standard deviations \( X \) is from the mean  
- Used for comparing scores from different distributions

---

## 17. Point and interval estimates

- **Point estimate**: Single best guess (e.g., sample mean)  
- **Interval estimate**: Range around the point (e.g., 95% confidence interval)

---

## 18. Significance of confidence intervals

- Indicates the range of values where the true population parameter is likely to lie  
- 95% CI: "We are 95% confident..."

---

## 19. Z-score and confidence interval relation

- CI formula:  
\[
\bar{X} \pm Z \cdot \frac{\sigma}{\sqrt{n}}
\]  
- Z-score defines margin of error  
  - 95% → Z = 1.96  
  - 99% → Z = 2.576

---

## 20. Comparing distributions with Z-scores

- Standardizes different distributions  
- Example: Compare a test score from two exams with different means and std devs

---

## 21. Assumptions for CLT

1. Independent random samples  
2. Identically distributed  
3. Finite variance and mean  
4. Large \( n \) (typically \( \geq 30 \))

---

## 22. What is expected value?

- The **mean** of a random variable  
- Discrete:  
\[
E(X) = \sum x_i \cdot P(x_i)
\]  
- Continuous:  
\[
E(X) = \int x \cdot f(x) dx
\]

---

## 23. Relation between probability distribution and expected outcome

- The **probability distribution** gives all possible outcomes and their likelihoods  
- The **expected value** is the long-run average from this distribution
